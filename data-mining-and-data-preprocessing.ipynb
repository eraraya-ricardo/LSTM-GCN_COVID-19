```python

df_national = pd.read_csv(filepath_or_buffer = 'E:\Tugas\SAR COV 19 in Indonesia Dataset\CSVs\prov_template.csv', 
                          index_col=0) #delete 'Unnamed: 0' column

for i in march: #iterate through each province in the dictionary
    with urllib.request.urlopen('https://data.covid19.go.id/public/api/prov_detail_' + province_code[i]+ '.json') as url: 
        data = json.loads(url.read().decode())#read json
    print('loaded '+province_code[i], i)
    #coloumns dictionary to store
    columns_dict = {
                    'tanggal': [], 
                    'KASUS': [], 
                    'MENINGGAL': [], 
                    'SEMBUH': []
                   }
    
    #convert dictionary into dataframe
    df = pd.DataFrame(columns_dict)
        
    ts = data['list_perkembangan'][0]['tanggal']/1000
    ts_conv = datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S') #convert to time string
    flipped_ts = ts_conv[8:10] #day
    flipped_ts = flipped_ts + ts_conv[4:8] #month
    flipped_ts = flipped_ts + ts_conv[0:4] #year
    case_zero = flipped_ts
    print(case_zero)
    #start date format
    dd = int(case_zero[0:2])
    for l in range(int(case_zero[0:2])-1):
        dd -= 1
        minus_one = str(dd) + '-03-2020'
        if len(minus_one) == 9:
            print('tanggal satuan')
            minus_one = '0' + minus_one
            print(minus_one)
        template_dict = {
                        'tanggal': minus_one, 
                        'KASUS': 0, 
                        'MENINGGAL': 0, 
                        'SEMBUH': 0
                        }
        template_df = pd.DataFrame(template_dict, index = [0])
        df = pd.concat([template_df, df])
    
    #insert new column with same value, the province name
    df.insert(loc=0, column='province', value=province_code[i])
    df['population'] = province_detail['population'][i]
    df['area'] = province_detail['area'][i]
    df['north'] = polygon_border['north'][i]
    df['south'] = polygon_border['south'][i]
    df['east'] = polygon_border['east'][i]
    df['west'] = polygon_border['west'][i]
    
    #rename the columns
    df.columns = ['province', 'date', 'case', 'death', 'recovered', 'population', 'area ($km^2$)', 
                  'north border (latitude)','south border (latitude)', 'east border (longitude)', 'west border (longitude)']
    #concat
    df_national = pd.concat([df_national, df])

    #clean up 
    df_national.reset_index(inplace = True) #reset index number
    df_national = df_national.drop(['index'], axis=1) #drop the unused index column

    #save dataframe as csv
    df_national.to_csv(path_or_buf = 'E:\Tugas\SAR COV 19 in Indonesia Dataset\CSVs\prov_template_' + province_code[i] + '.csv')
    #report
    print(province_code[i] + '.csv saved')
```
