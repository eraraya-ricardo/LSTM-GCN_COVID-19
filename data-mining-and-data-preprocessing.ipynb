# Run to Update

Update Every Province CSV File

```python
#list for logging df length later
prov_len = []

for i in range(34): #iterate through each province in the dictionary
    with urllib.request.urlopen('https://data.covid19.go.id/public/api/prov_detail_' + province_code[i]+ '.json') as url: 
        data = json.loads(url.read().decode())#read json
    
    #coloumns dictionary to store
    columns_dict = {
                    'tanggal': [], 
                    'KASUS': [], 
                    'MENINGGAL': [], 
                    'SEMBUH': []
                   }

    for item in data['list_perkembangan']: #iterate through each date
            for column in columns_dict.keys(): #iterate through each column
                    columns_dict[column].append(item[column]) #add spesific data to the corresponding column
   
    
    #convert dictionary into dataframe
    df = pd.DataFrame(columns_dict)
        
    #date converter
    df['tanggal'] = df['tanggal']/1000 #delete time detail on UNIX Time
    for j in range(len(data['list_perkembangan'])): #iterate through each date
        ts = df['tanggal'][j] 
        ts_conv = datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S') #convert to time string
        flipped_ts = ts_conv[8:10] #day
        flipped_ts = flipped_ts + ts_conv[4:8] #month
        flipped_ts = flipped_ts + ts_conv[0:4] #year
        df['tanggal'][j] = flipped_ts
    
    #insert new columns with same value
    df.insert(loc=0, column='province', value=province_code[i])
    df.insert(loc=3, column='confirmed (cumulative)', value=df['KASUS'].cumsum())
    df.insert(loc=5, column='deaths (cumulative)', value=df['MENINGGAL'].cumsum())
    df.insert(loc=7, column='recovered (cumulative)', value=df['SEMBUH'].cumsum())
    df['population'] = province_detail['population'][i]
    df['area'] = province_detail['area'][i]
    df['north'] = polygon_border['north'][i]
    df['south'] = polygon_border['south'][i]
    df['east'] = polygon_border['east'][i]
    df['west'] = polygon_border['west'][i]
    df['capital'] = capital_code[i]
    df['capital_lat'] = capital_coordinate['lat'][i]
    df['capital_lon'] = capital_coordinate['lon'][i]
        
    #rename columns
    df.columns = ['province', 'date', 'confirmed', 'confirmed (cumulative)', 'deaths', 'deaths (cumulative)', 'recovered', 'recovered (cumulative)', 
                  'population', 'area ($km^2$)', 'north border (latitude)','south border (latitude)', 'east border (longitude)',
                  'west border (longitude)', 'capital', 'capital coordinate (latitude)', 'capital coordinate (longitude)']
    
    #read province template csv
    df_template = pd.read_csv(filepath_or_buffer = 
                              '/content/drive/My Drive/Programming-Project/SAR-COV-19/LSTM-GCN_COVID-19/assets/prov_templates/prov_template_' 
                              + province_code[i] + '.csv')
    #clean up
    df_template = df_template.drop(['Unnamed: 0'], axis=1) #drop the unused index column
    
    #concat
    df = pd.concat([df_template, df])

    #clean up 
    df.reset_index(inplace = True) #reset index number
    df = df.drop(['index'], axis=1) #drop the unused index column

    #logging df length for concatenating all province later
    prov_len.append(len(df))

    #save dataframe as csv
    df.to_csv(path_or_buf = '/content/drive/My Drive/Programming-Project/SAR-COV-19/LSTM-GCN_COVID-19/assets/prov_details/prov_detail_' 
                      + province_code[i] + '.csv')
    #report
    print(i, 'prov_detail_' + province_code[i] + '.csv saved')
print('done')
```
